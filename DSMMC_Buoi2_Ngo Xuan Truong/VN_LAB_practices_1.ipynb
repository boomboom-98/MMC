## Question 3: Viết chương trình liệt kê tất cả các số nguyên tố nhỏ hơn n cho trước.


```python
a3 = int(input("Nhập n: "))
dem3 = [1]
for i in range(2,a3+1): #Duyệt các số từ 1 đến n xem có là số nguyên tố không
  for j in range(2,i):  #Tìm ước của từng số
    if i % j == 0:
      break
    else:
      dem3.append(i)

print(dem3)
```
